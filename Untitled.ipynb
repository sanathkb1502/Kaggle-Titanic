```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df.shape
```

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace = True)
df
```

```python
df.columns
```

```python
def print_correlation_list(df):
    for i in df.columns:
        if (df[i].dtype != object):
            print(i,"vs survived", df[i].corr(df['Survived']))
```

```python
def Embarked_value(embark_place):
    ar = ['C','Q','S']
    return ar.index(embark_place)
    
```

```python
featuresList = ["Pclass", "Age", "SibSp", "Parch", "Fare", "Survived"]
features = pd.DataFrame(columns = featuresList)
for i in featuresList:
    features[i] = df[i]

features["Sex"] = df['Sex'].apply(lambda x: 1 if x.lower() == 'male' else 0)
features["embark_value"] = df['Embarked'].apply(lambda x : Embarked_value(x))
features
```

```python
df.dropna(subset=['Embarked'],inplace = True)
df.shape
```

```python
for i in features.columns:
    if (features[i].dtype != object):
        print(i,"vs survived", features[i].corr(features['Survived']))
```

```python
print_correlation_list(features)
```
